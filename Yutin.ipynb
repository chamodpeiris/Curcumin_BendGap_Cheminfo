```python
import pandas as pd

yuting = pd.read_csv('https://raw.githubusercontent.com/AjStephan/Yuting/main/ML%20data.csv')

yuting = yuting[yuting['reaction time'] == 120]
# Rename columns to match training data
yuting = yuting.rename(columns={
    "metal/%": "metal_",
    "calcination temperature": "calcination_temp",
    "calcination time": "calcination_time",
})


# Metal HOMO and LUMO values
metal_homo = {
    'MnO2': -0.487, 'Ag': -1.868, 'Co': -1.543, 'none': -2.499,
    'Fe': -0.288, 'Cu': -1.051, 'Ni': -0.297}
metal_lumo = {
    'MnO2': 0.273, 'Ag': -0.887, 'Co': -0.83, 'none': -0.625,
    'Fe': 0.15, 'Cu': 0.139, 'Ni': 0.149}

# Replace metal names with their corresponding values
yuting['metal_homo'] = yuting['metal'].map(metal_homo)
yuting['metal_lumo'] = yuting['metal'].map(metal_lumo)
yuting['band_gap'] = yuting['metal_homo'] - yuting['metal_lumo']
yuting = yuting[yuting['metal'] != 'TiO2']
yuting = yuting.drop(['Z', 'Z rad', 'E neg', 'Ion eng', 'MM', 'Redox'], axis=1)
yuting = yuting.drop(['metal'], axis=1)
yuting = yuting.reset_index(drop=True)
yuting.head()
```
