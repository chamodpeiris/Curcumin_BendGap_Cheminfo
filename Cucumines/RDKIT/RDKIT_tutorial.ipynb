```python
# Import necessary modules
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import Descriptors

# Load a molecule
smiles = "CCCCCCc1ccccc1"
mol = Chem.MolFromSmiles(smiles)

# Draw the molecule
Draw.MolToImage(mol)
```

```python
# Calculate molecular descriptors
mw = Descriptors.MolWt(mol)
logp = Descriptors.MolLogP(mol)
tpsa = Descriptors.TPSA(mol)
bandgap = Descriptors.NumValenceElectrons(mol)

# Print the calculated descriptors
print(f'Molecular weight: {mw}')
print(f'LogP: {logp}')
print(f'TPSA: {tpsa}')
print(f'Band gap: {bandgap}')
```

```python
import pycaret.regression as pyreg
from pycaret.datasets import get_data
import matplotlib.pyplot as plt

# Load dataset
data = get_data('boston')

# Initialize the regression setup
reg = pyreg.setup(data, target='medv', html=False)

# Compare models
best_model = pyreg.compare_models()

# Pull the results
compare_df = pyreg.pull()

# Plot R^2 scores
plt.figure(figsize=(12, 6))
plt.barh(compare_df.index, compare_df['R2'])
plt.xlabel('R^2 Score')
plt.title('Comparison of Models based on R^2 Score')
plt.grid(True)
plt.show()
```
